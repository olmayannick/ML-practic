```python
import yfinance as yf   
```

```python
google = yf.Ticker("GOOGL")
df= google.history(period="1d", start = "2023-01-01", end = "2023-12-31")
df
```

```python
df.plot(y="Close")
```

```python
df.plot(y="Open")
```

```python
apple = yf.Ticker("AAPL")
df= apple.history(period="1d", start = "2020-01-01", end = "2023-12-31")
df.plot(y="Close")
```

```python
tesla = yf.Ticker("TSLA")
df= tesla.history(period="1d", start = "2020-01-01", end = "2023-12-31")
df.plot(y="Close")
```

```python
df=yf.download(["TSLA","AAPL","MSTF","AKBNK.IS"],start="2024-1-1", end="2024-2-29")
df.plot(y="Open")
```
